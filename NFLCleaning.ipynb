```
nfl_classification_data['game_half'] = nfl_classification_data.game_half.apply(lambda row: 
                                                                               0 if row == 'Half1'
                                                                               else 1 if row == 'Half2'
                                                                               else 2 if row == 'Overtime'
                                                                               else 3)

nfl_classification_data['pass_length'] = nfl_classification_data.pass_length.apply(lambda row:
                                                                                   0 if row == 'short'
                                                                                   else 1 if row == 'deep'
                                                                                   else 2)

nfl_classification_data['run_location'] = nfl_classification_data.run_location.apply(lambda row:
                                                                                    0 if row == 'right'
                                                                                    else 1 if row == 'left'
                                                                                    else 2 if row == 'middle'
                                                                                    else 3)
```

```
nfl_classification_data['pass_length'] = nfl_classification_data.pass_length[nfl_classification_data.pass_length < 2]
```

```
nfl_classification_data['run_location'] = nfl_classification_data.run_location[nfl_classification_data.run_location < 3]
```
